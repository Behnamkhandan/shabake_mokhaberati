```python
import numpy as np
```

```python
def crc_encode(data, generator):
    data = data + '0'*(len(generator)-1)   # اضافه کردن صفرها
    data = list(data)
    generator = list(generator)
    
    for i in range(len(data)-len(generator)+1):
        if data[i] == '1':
            for j in range(len(generator)):
                data[i+j] = str(int(data[i+j] != generator[j]))  # XOR
    
    return ''.join(data[-(len(generator)-1):])  # باقی‌مانده CRC
```

```python
def crc_check(data_with_crc, generator):
    data = list(data_with_crc)
    generator = list(generator)
    
    for i in range(len(data)-len(generator)+1):
        if data[i] == '1':
            for j in range(len(generator)):
                data[i+j] = str(int(data[i+j] != generator[j]))
    
    remainder = ''.join(data[-(len(generator)-1):])
    return remainder == '0'*(len(generator)-1)
```

```python
# داده و پلی‌نوم ژنراتور
data = "1101011011"
generator = "10011"

# محاسبه CRC
crc = crc_encode(data, generator)
print("CRC:", crc)

# داده همراه CRC
data_with_crc = data + crc
print("Data with CRC:", data_with_crc)

# بررسی صحت
print("Check (no error):", crc_check(data_with_crc, generator))

# شبیه‌سازی خطا
data_with_error = data_with_crc[:-1] + ('1' if data_with_crc[-1]=='0' else '0')
print("Check (with error):", crc_check(data_with_error, generator))
```
